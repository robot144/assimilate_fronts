# Distance based cost functions

In data-assimilation, we normally compare the observations $y$ to the corresponding model output $H(x)$ as $y-H(x)$, but this is known to perform poorly to spatially align features with sharp transitions. This notebook explores some alternatives in a simple 1d context. One can think of rain versus no rain or clouds that can be considered as present or not in a particular pixel. Normally, we mostly consider data with multiple dimensions, but here we start with a single spatial dimension for easy analysis and visualization.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
x=np.arange(0.0,10.0,0.1)
```

## Description of the problem

Consider a simple dataset, where the observations tell us that it rains between $4<x<6$. Rain is indicated by value one and no-rain by value zero. The initial model values have a spatial offset of 1 unit (eg km)


```python
def case1(x,offset=0.0):
    result=np.zeros(x.shape)
    result[((x-offset)>=4.0) & ((x-offset)<6.0)]=1.0
    return result

y1=case1(x)
h1=case1(x,1.0)
```

```python
line1=plt.plot(x,y1,'k-',label='measurement y')
line2=plt.plot(x,h1,'b-',label='model H(x)')
plt.legend()
plt.show()

plt.plot(x,y1-h1,'g-', label='y-H(x)')
plt.title('Blocky residual for sharp transitions')
plt.legend()
plt.show()
```
